## ¿Cómo afecta la contaminación del aire a la salud en España? Un análisis integrado de calidad del aire y hospitalizaciones

# Objetivo:
El propósito de este análisis es evaluar cómo los niveles de contaminación atmosférica en España se relacionan con indicadores de salud pública, especialmente hospitalizaciones y mortalidad por causas respiratorias. Para ello se integran datos oficiales de calidad del aire, población y salud, con el fin de identificar patrones, riesgos y posibles desigualdades territoriales.
- Analizar la relación entre contaminantes atmosféricos y hospitalizaciones/mortalidad respiratoria en España, identificando patrones temporales, geográficos y demográficos.

# Motivación
La contaminación atmosférica es uno de los principales riesgos ambientales para la salud humana. Según organismos internacionales, la exposición prolongada a partículas finas (PM2.5), dióxido de nitrógeno (NO₂) y ozono (O₃) incrementa el riesgo de enfermedades respiratorias, cardiovasculares y mortalidad prematura.
En España, aunque la calidad del aire ha mejorado en la última década, persisten episodios de alta contaminación en áreas urbanas e industriales. Sin embargo, no siempre está claro cómo estos niveles se traducen en impactos reales sobre la salud, ni si todas las regiones se ven afectadas por igual.


# Preguntas que guían el análisis
- ¿Qué contaminantes presentan mayores niveles en España y cómo varían por región y estación del año?
- ¿Existen correlaciones claras entre picos de contaminación y aumentos en hospitalizaciones respiratorias?
- ¿Qué ciudades o comunidades autónomas muestran mayor vulnerabilidad?
- ¿Afectan igual los contaminantes a todas las edades o grupos poblacionales?
- ¿Se observan tendencias preocupantes (incrementos, estacionalidad, episodios extremos)?
- ¿Qué contaminantes parecen tener mayor impacto en la salud?


# Hipótesis del proyecto
- Hipótesis 1: PM2.5 y NO₂ son los contaminantes más asociados a hospitalizaciones respiratorias.
Basado en evidencia científica previa, se espera que estos contaminantes muestren correlaciones más fuertes con indicadores de salud.
- Hipótesis 2: Las grandes ciudades presentan peores niveles de contaminación y mayor carga hospitalaria.
Madrid, Barcelona, Valencia o Bilbao podrían mostrar patrones más marcados debido al tráfico y la densidad poblacional.
- Hipótesis 3: Existe estacionalidad clara: más contaminación en invierno y más ozono en verano.
Esto podría reflejarse en picos de hospitalizaciones en determinadas épocas del año.
- Hipótesis 4: Las zonas con población más envejecida presentan mayor vulnerabilidad.
La combinación de contaminación + envejecimiento podría amplificar los efectos en salud.
- Hipótesis 5: La contaminación no afecta por igual a todas las regiones.
Se espera encontrar desigualdades territoriales relacionadas con industria, tráfico, clima y orografía.
